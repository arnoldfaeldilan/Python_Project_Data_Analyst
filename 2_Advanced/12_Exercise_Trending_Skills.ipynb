```python
import pandas as pd
import matplotlib.pyplot as plt
import ast

# load dataset
df = pd.read_csv('C:/Users/HP/Desktop/Python_Data_Project/datasets/data_jobs.csv')

# data cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(
    lambda skill: ast.literal_eval(skill)
    if isinstance(skill, str)
    else skill
)
```

```python
job_title = 'Data Analyst'
df_data_analyst = df[df['job_title_short'] == job_title].copy()
df_data_analyst
```

```python
df_data_analyst['job_posted_month'] = df['job_posted_date'].dt.month
df_data_analyst
```

```python
df_data_analyst_exploded = df_data_analyst.explode('job_skills')
df_data_analyst_exploded
```

```python
df_data_analyst_exploded_pivot = df_data_analyst_exploded.pivot_table(
    index='job_posted_month',
    columns='job_skills',
    aggfunc='size',
    fill_value=0
)
df_data_analyst_exploded_pivot
```

```python
df_data_analyst_exploded_pivot.loc['Total'] = df_data_analyst_exploded_pivot.sum()
df_data_analyst_exploded_pivot
```

```python
df_DA = df_data_analyst_exploded_pivot[df_data_analyst_exploded_pivot.loc['Total'].sort_values(ascending=False).index].drop('Total')
df_DA
```

```python
df_DA.iloc[:, :5]
```

```python
df_DA.iloc[:, :5].plot(kind='line')
plt.title(f'Top 5 Skills for {job_title}s per Month')
plt.xlabel('Month')
plt.ylim(2500, 12000)
plt.legend()
```

```python
df_data_jobs = df.copy()
title_job = 'Data Analyst'
df_data_jobs_copy = df_data_jobs[df_data_jobs['job_title_short'] == title_job].copy()
df_data_jobs_copy['job_posted_month'] = df_data_jobs_copy['job_posted_date'].dt.strftime('%b')
df_data_jobs_copy
```

Explode the job_skills column
Pivot Table

```python
df_data_jobs_copy_ex = df_data_jobs_copy.explode('job_skills')
df_data_jobs_copy_ex_piv = df_data_jobs_copy_ex.pivot_table(
    index='job_posted_month',
    columns='job_skills',
    aggfunc='size',
    fill_value=0
)
df_data_jobs_copy_ex_piv
```

```python
# create a column
df_data_jobs_copy_ex_piv.reset_index(inplace=True)
df_data_jobs_copy_ex_piv['month_no'] = pd.to_datetime(df_data_jobs_copy_ex_piv['job_posted_month'], format='%b').dt.month
df_data_jobs_copy_ex_piv
```

```python
df_data_jobs_copy_ex_piv.sort_values(by='month_no', inplace=True)
df_data_jobs_copy_ex_piv.set_index('job_posted_month', inplace=True)
df_data_jobs_copy_ex_piv.drop(columns='month_no', inplace=True)
df_data_jobs_copy_ex_piv
```

```python
top_skills = 5
top_skills_list = df_data_jobs_copy_ex_piv.sum().sort_values(ascending=False).head(top_skills).index.to_list()
top_skills_list
```

```python
df_data_jobs_copy_ex_piv[top_skills_list].plot(kind='line')
plt.title(f'Top {top_skills} Skills for {title_job}s per Month')
plt.xlabel('Month')
plt.show()
```
