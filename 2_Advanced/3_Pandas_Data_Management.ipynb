# ```Copy()```

Copy() creates a copy of a dataframe

````python
# load a new datasets
df_original = pd.read_csv('C:/Users/HP/Desktop/Python_Data_Project/datasets/data_jobs.csv')
````

````python
# create a copy of a data frame
df_altered = df_original.copy()
````

````python
# checks the identity
print(f"{'ID of df_original:':30} {id(df_original)}")
print(f"{'ID of df_altered:':30} {id(df_altered)}")
print(f"{'Are the Dataframees the same?':30} {id(df_original)==id(df_altered)}")
````

````python
# median salary of df_altered (copy of df_original)
median_salary_altered = df_altered['salary_year_avg'].median()

df_altered['salary_year_avg'] = df_altered['salary_year_avg'].fillna(median_salary_altered)

df_altered.loc[0:5, 'salary_year_avg']
````
