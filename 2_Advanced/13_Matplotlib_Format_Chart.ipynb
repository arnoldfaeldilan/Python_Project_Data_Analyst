# Matplotlib Formatting Charts

```python
import pandas as pd
import matplotlib.pyplot as plt
import ast

# load dataset
df = pd.read_csv('C:/Users/HP/Desktop/Python_Data_Project/datasets/data_jobs.csv')

# data cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else x)
```

Job Count and Job Schedule Type Count

```python
df['job_title_short'].value_counts()
```

```python
df['job_schedule_type'].value_counts().head(3)
```

```python
fig, ax = plt.subplots(1, 2)

df['job_title_short'].value_counts().plot(kind='bar', ax=ax[1])
df['job_schedule_type'].value_counts().head(3).plot(kind='bar', ax=ax[0])

fig.tight_layout() # to prevent values overlapping
```

```python
fig, ax = plt.subplots(2, 1)

df['job_title_short'].value_counts().sort_values(ascending=True).plot(kind='barh', ax=ax[1])
df['job_schedule_type'].value_counts().head(3).plot(kind='barh', ax=ax[0])

fig.tight_layout() # to prevent values overlapping
```

# Count of Top Skills in Job Postings

```python
df_skills = df.copy()
df_skills = df_skills.explode('job_skills')
skill_counts  = df_skills.groupby(['job_skills', 'job_title_short']).size()
df_skills_count = skill_counts.reset_index(name='skill_count')
df_skills_count.sort_values(by='skill_count', ascending=False, inplace=True)
df_skills_count
```

```python
job_titles = ['Data Scientist','Data Engineer','Data Analyst']
```

```python
fig, ax = plt.subplots(3,1)

for i, job_title in enumerate(job_titles):
    df_plot = df_skills_count[df_skills_count['job_title_short'] == job_title].head(5)
    df_plot.plot(kind='barh', x='job_skills', y='skill_count', ax=ax[i], title=job_title)
    ax[i].invert_yaxis()
    ax[i].legend().set_visible(False)
    ax[i].set_ylabel('')
    ax[i].set_xlim(0, 120000)

fig.suptitle('Counts of Top skills in Job Posting', fontsize=15)
fig.tight_layout()
```
