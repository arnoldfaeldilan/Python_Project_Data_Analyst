```python
import pandas as pd
import matplotlib.pyplot as plt
import ast

# load dataset
df = pd.read_csv('C:/Users/HP/Desktop/Python_Data_Project/datasets/data_jobs.csv')

# data cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda skills: ast.literal_eval(skills) if isinstance(skills, str) else skills)
```

```python
df.explode('job_skills')
```

```python
# explode the job_skills
df.explode('job_skills')
```

```python
# count the job_skills
df.explode('job_skills').value_counts('job_skills')
```

```python
# select only the top 10
top_10_skills = df.explode('job_skills').value_counts('job_skills').head(10)

top_10_skills
```

```python
top_10_skills = top_10_skills.sort_values(ascending=True)

top_10_skills
```

```python
# plot
top_10_skills.plot(kind='barh')
```

## Generate a Bar Graph of top skills in Job title

```python
df_exploded = df.explode('job_skills')

df_exploded
```

```python
job_skill_count = df_exploded.groupby(['job_title_short', 'job_skills']).size()

job_skill_count
```

```python
df_job_skills = job_skill_count.reset_index(name = 'job_count').sort_values(by = 'job_count', ascending=False)

df_job_skills
```

```python
job_title = 'Data Engineer'
top_skills = 15

df_final = df_job_skills[df_job_skills['job_title_short'] == job_title].head(top_skills)

df_final.plot(kind = 'barh', x = 'job_skills', y = 'job_count')
plt.gca().invert_yaxis() 
plt.title(f'The Top {top_skills} Skills for {job_title}s')
plt.xlabel('Job Posting Count')
plt.ylabel('')
plt.legend().set_visible(False)
plt.show()
```
