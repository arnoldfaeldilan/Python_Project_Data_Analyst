```python
import pandas as pd
import matplotlib.pyplot as plt

# loading datasets
df = pd.read_csv('C:/Users/HP/Desktop/Python_Data_Project/datasets/data_jobs.csv')

# data cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df.info()
```

```python
median_year_avg = df['salary_year_avg'].median()
```

```python
median_hour_avg = df['salary_hour_avg'].median()
```

```python
# we will replace all none and nan values with these median values
# fillna() method is to fill missing values (NaN)
df_filled = df

df_filled['salary_year_avg'] = df_filled['salary_year_avg'].fillna(median_year_avg)
df_filled['salary_hour_avg'] = df_filled['salary_hour_avg'].fillna(median_hour_avg)
```

```python
df_filled.loc[0:10, 'salary_year_avg':'salary_hour_avg']
```

```python
# drop duplicates

df_unique = df_filled
df_unique = df_unique.drop_duplicates()

# to know how many duplicates were dropped
print(f"{'Length of original df:':30} {len(df_filled)}")
print(f"{'Length after drop duplicates:':30} {len(df_unique)}")
print(f"{'Rows Dropped:':30} {len(df_filled) - len(df_unique)}")
```

```python
# drop duplicate from repetitive job title from same company

df_unique = df_unique.drop_duplicates(subset=['job_title', 'company_name'])

# to know how many duplicates were dropped
print(f"{'Length of original df:':30} {len(df_filled)}")
print(f"{'Length after drop duplicates:':30} {len(df_unique)}")
print(f"{'Rows Dropped:':30} {len(df_filled) - len(df_unique)}")
```
