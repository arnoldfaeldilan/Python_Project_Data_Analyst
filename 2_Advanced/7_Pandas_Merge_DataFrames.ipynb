````python
import matplotlib.pyplot as plt

# load dataset
df = pd.read_csv('C:/Users/HP/Desktop/Python_Data_Project/datasets/data_jobs.csv')

# data cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
````

````python
# create a new dataframe with only united states 
df_usa = df[df['job_country'] == 'United States'].copy()

df_usa
````

````python
df_usa['job_posted_month'] = df_usa['job_posted_date'].dt.strftime('%B')

df_usa['job_posted_month']
````

````python
df_usa_pivot = df_usa.pivot_table(index='job_posted_month', columns='job_title_short', aggfunc='size')

df_usa_pivot
````

````python
df_usa_pivot.reset_index(inplace=True)
df_usa_pivot['month_no'] = pd.to_datetime(df_usa_pivot['job_posted_month'], format='%B').dt.month

df_usa_pivot.sort_values('month_no', inplace=True)
df_usa_pivot.set_index('job_posted_month', inplace=True)
df_usa_pivot.drop(columns='month_no', inplace=True)
df_usa_pivot
````

````python
df_us_software_jobs = pd.read_csv('C:/Users/HP/Desktop/Python_Data_Project/datasets/software_jobs.csv')

df_us_software_jobs.set_index('job_posted_month', inplace=True)

df_us_software_jobs
````

````python
df_us_software_merged = df_usa_pivot.merge(df_us_software_jobs, how='inner',on = 'job_posted_month')

df_us_software_merged
````

````python
df_us_software_merged.plot(kind='line')
````

````python
top_5_jobs = (
    df_us_software_merged
    .sum()
    .sort_values(ascending=False)
    .head(5)
    .index # get the index of top 5
    .to_list() # conver the index into list
    )

top_5_jobs
````

````python
df_us_software_merged[top_5_jobs].plot(kind='line')
plt.title('Monthly Job Posting for Top Tech Jobs in the US')
plt.xlabel('2023')
plt.ylabel('Job Count')
plt.ylim(0, 20000)
plt.legend()
plt.show()
````
